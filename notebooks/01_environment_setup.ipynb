```python
import torch, transformers
print("transformers", transformers.__version__)   # want 4.43.4

from transformers import AutoTokenizer, AutoModel
from transformers.models.bert.configuration_bert import BertConfig

name   = "zhihan1996/DNABERT-2-117M"
config = BertConfig.from_pretrained(name)
tok    = AutoTokenizer.from_pretrained(name, trust_remote_code=True)
model  = AutoModel.from_pretrained(name, trust_remote_code=True, config=config).eval()

seq = "ACGTAGCATCGGATCTATCTATCGACACTTGGTTATCGATCTACGAGCATCTCGTTAGC"
ids = tok(seq, return_tensors="pt")["input_ids"]      # CPU is fine
with torch.no_grad():
    hidden = model(ids)[0]        # last_hidden_state: [1, n_tokens, 768]
print("hidden", tuple(hidden.shape), "mean", float(hidden.mean()))
```
